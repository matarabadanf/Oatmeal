# Bound state core ionization of atoms using CS-RHF

```python
import numpy as np
from py_mods.src.SCF.CSUHF import CS_UHF
from py_mods.src.SCF.external import UHF_context_from_pyscf
```

## $He$ $1s2s$ and $He^{-}$ $1s2s^{2}$

```python
He_1s2s_context = UHF_context_from_pyscf("He 0 0 0", 0, 0, basis="aug-cc-pvqz")
He_1s2s_context.mult = 2
He_1s2s_context.p_guess = 'RHF'
He_1s2s_context.occupation = (np.array([1, 1]), np.array([0, 0]))
# He_1s2s_context.verbose = True
```

```python
He_1s2s_res = CS_UHF(He_1s2s_context)
E_ref_He_min_1s2s = -2.1743
print(
    f"${He_1s2s_res.E_UHF.real:.4f}$ | ${He_1s2s_res.E_UHF.real - E_ref_He_min_1s2s :.4f}$ |"
)
```

```python
He_min_1s2s_context = UHF_context_from_pyscf("He 0 0 0", 0, 0, basis="aug-cc-pvqz")
He_min_1s2s_context.mult = 1
He_min_1s2s_context.n_electrons = 3
He_min_1s2s_context.conv_type = None
He_min_1s2s_context.p_guess = 'RHF'
He_min_1s2s_context.occupation = (np.array([0, 1]), np.array([1, 1]))
He_min_1s2s_context.verbose = True
```

```python
He_min_1s2s_res = CS_UHF(He_min_1s2s_context)
E_ref_He_min_1s2s2 = -2.1633
print(
    f"${He_min_1s2s_res.E_UHF.real:.4f}$ | ${He_min_1s2s_res.E_UHF.real - E_ref_He_min_1s2s2 :.4f}$ |"
)
```

## $Ne$ $2p^{5}3s$ and $Ne^-$ $2p^{5}3s^{2}$

```python
Ne_2p5_3s_context = UHF_context_from_pyscf("Ne 0 0 0", 0, 0, basis="aug-cc-pvqz")
Ne_2p5_3s_context.mult = 2
Ne_2p5_3s_context.p_guess = 'RHF'
Ne_2p5_3s_context.occupation = (np.array([1, 1, 1, 1, 1]), np.array([1, 1, 1, 1, 0, 1]))
# Ne_2p5_3s_context.verbose = True
```

```python
Ne_2p5_3s_res = CS_UHF(Ne_2p5_3s_context)
E_ref_Ne_2p5_3s = -127.9923
print(
    f"${Ne_2p5_3s_res.E_UHF.real:.4f}$ | ${Ne_2p5_3s_res.E_UHF.real - E_ref_Ne_2p5_3s :.4f}$ |"
)
```

```python
Ne_min_2p5_3s2_context = UHF_context_from_pyscf("Ne 0 0 0", 0, 0, basis="aug-cc-pvqz")
Ne_min_2p5_3s2_context.mult = 1
Ne_min_2p5_3s2_context.occupation = (
    np.array([1, 1, 1, 1, 1, 1]),
    np.array([1, 1, 1, 1, 0, 1]),
)
Ne_min_2p5_3s2_context.p_guess = 'RHF'
Ne_min_2p5_3s2_context.break_symm = True
Ne_min_2p5_3s2_context.verbose = True
Ne_min_2p5_3s2_context.conv_type =  'DIIS'
Ne_min_2p5_3s2_context.n_electrons = 11 
Ne_min_2p5_3s2_context.max_iter = 110
Ne_min_2p5_3s2_context.acc_iteration_start =  100
```

```python
Ne_min_2p5_3s2_res = CS_UHF(Ne_min_2p5_3s2_context)
E_ref_Ne_min_2p5_3s2 = -127.9867
print(
    f"${Ne_min_2p5_3s2_res.E_UHF.real:.4f}$ | ${Ne_min_2p5_3s2_res.E_UHF.real - E_ref_Ne_min_2p5_3s2 :.4f}$ |"
)
```
